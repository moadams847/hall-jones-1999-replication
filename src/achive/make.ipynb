```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
from scipy import stats

# --- Settings for 1280 x 720 px ---
WIDTH_PX, HEIGHT_PX = 1280, 720
DPI = 160  # choose 100/120/150/160; must use the same in savefig
WIDTH_IN, HEIGHT_IN = WIDTH_PX / DPI, HEIGHT_PX / DPI  # -> (8.0, 4.5) at 160 dpi

# ── ONLY CHANGE 1: read the extension (2019) data instead of replication ──
df = pd.read_csv(r'C:\Users\Adams\OneDrive\DE & E Research\outputs\sample_B_replication_extended.csv')

df['log_yl'] = np.log(df['yl_adj'])
df['S'] = df['social_infra']
df['imputed'] = df['si_was_imputed'].astype(bool)

obs = df[~df['imputed']]
imp = df[df['imputed']]

N = len(df)
N_obs = len(obs)
N_imp = len(imp)

slope, intercept, r_value, p_value, se = stats.linregress(df['S'], df['log_yl'])
r2 = r_value**2

# confidence band
x_fit = np.linspace(df['S'].min() - 0.02, df['S'].max() + 0.02, 300)
y_fit = intercept + slope * x_fit

n = len(df)
x_mean = df['S'].mean()
s_xx = np.sum((df['S'] - x_mean)**2)
se_fit = se * np.sqrt(1/n + (x_fit - x_mean)**2 / s_xx)
t_crit = stats.t.ppf(0.975, df=n-2)
y_upper = y_fit + t_crit * se_fit
y_lower = y_fit - t_crit * se_fit

# stars
if p_value < 0.001:
    stars = '***'
elif p_value < 0.01:
    stars = '**'
elif p_value < 0.05:
    stars = '*'
else:
    stars = ''

# ---- FIGURE: exact 1280x720 when saved ----
fig, ax = plt.subplots(figsize=(WIDTH_IN, HEIGHT_IN), dpi=DPI, constrained_layout=True)
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# confidence band
ax.fill_between(x_fit, y_lower, y_upper, color='#4a9b8e', alpha=0.18, zorder=1)

# OLS line
ax.plot(x_fit, y_fit, color='#2d6e65', linewidth=2.2, zorder=3,
        label=f'OLS fit:  log(Y/L) = {intercept:.2f} + {slope:.2f} · S')

# scatter
ax.scatter(obs['S'], obs['log_yl'], color='#1a2e3a', s=22, zorder=4, marker='o',
           label=f'Observed S  (N={N_obs})')
ax.scatter(imp['S'], imp['log_yl'], color='#d4a017', s=22, zorder=4, marker='o',
           label=f'Imputed S   (N={N_imp})')

# labels
for _, row in df.iterrows():
    color = '#d4a017' if row['imputed'] else '#1a2e3a'
    ax.annotate(row['iso3'], (row['S'], row['log_yl']),
                fontsize=6.5, color=color,
                xytext=(3, 1), textcoords='offset points',
                fontfamily='monospace')

# stats box
stats_text = f"$R^2$ = {r2:.3f}\nSlope = {slope:.3f}{stars}\n$N$ = {N}"
ax.text(0.02, 0.97, stats_text, transform=ax.transAxes,
        fontsize=10.5, verticalalignment='top',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='#cccccc', alpha=0.9),
        fontfamily='serif')

ax.set_xlabel('Social Infrastructure  (S)', fontsize=13, labelpad=8)
ax.set_ylabel('log(Output per Worker)', fontsize=13, labelpad=8)

# ── ONLY CHANGE 2: title updated to Extension (2019) ──────────────────────
ax.set_title(
    'Social Infrastructure vs. Output per Worker\n'
    'Extension (Recent Data) ·  $\\it{Gold}$ = imputed $S$,  $\\it{Navy}$ = observed $S$',
    fontsize=13.5, fontweight='bold', pad=12
)

ax.legend(loc='lower right', fontsize=9.5, frameon=True,
          framealpha=0.9, edgecolor='#cccccc')

ax.grid(True, linestyle='--', alpha=0.25, color='gray')
ax.tick_params(labelsize=10)

# ── ONLY CHANGE 3: output filename updated ───────────────────────────────
plt.savefig(
    r'C:\Users\Adams\OneDrive\DE & E Research\outputs\fig_yl_vs_S_extension_sampleB_New.png',
    dpi=DPI,
    bbox_inches=None,
    facecolor='white',
    pad_inches=0.05
)
print("Done")
```
